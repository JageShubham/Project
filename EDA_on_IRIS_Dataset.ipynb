#### Step 1: Import Libraries


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
from sklearn.datasets import load_iris
```

#### Step 2: Load the Iris Dataset

```python
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['target'] = iris.target
```

#### Step 4: Data Visualization

```python
# Pairplot for Overall Visualization
sns.pairplot(df, hue='target', palette='viridis')
plt.show()
```

**Observation**


> The pairplot displays relationships between different features, highlighting distinct clusters for each target class


> All type of flowers are well separable for PetalLenght and PetalWidth








```python
# Correlation Matrix Heatmap
plt.figure(figsize=(10, 5))
sns.heatmap(df1.corr(), annot=True, cmap='viridis', linewidths=.5)
plt.show()
```

**Observation**

> Sepal length and petal length, as well as petal length and petal width, exhibit particularly high positive correlations.

> PetalLenght and PetalWith have highest correlation i.e 0.96


> PetalLenght and SepalLenght have second highest correlation i.e 0.87










```python
# Boxplots for Individual Features
plt.figure(figsize=(10, 5))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i+1)
    colors = ["#FF9999", "#66B2FF", "#99FF99"]
    sns.boxplot(x='target', y=feature, data=df, palette=colors)
    plt.title(f'Boxplot of {feature}')
plt.tight_layout()
plt.show()
```

**Observation**


> Boxplots show the distribution of individual features for each target class.


> Sepal length and petal length have noticeable variations across different species.






```python
# Distribution of Features
plt.figure(figsize=(10, 5))
for i, feature in enumerate(iris.feature_names):
    plt.subplot(2, 2, i+1)
    sns.histplot(df, x=feature, bins=20, kde=True, hue='target', palette='viridis')
    plt.title(f'Distribution of {feature}')
plt.tight_layout()
plt.show()
```


**Observation**

> Histograms illustrate the distributions of sepal length, sepal width, petal length, and petal width.


#### Step 5: Conclusion


> The EDA provides a holistic understanding of the Iris dataset, showcasing distinct patterns and variations between different species.

> The dataset is relatively clean with no apparent outliers.


> There is strong correlation between PetalLenght and PetalWith
